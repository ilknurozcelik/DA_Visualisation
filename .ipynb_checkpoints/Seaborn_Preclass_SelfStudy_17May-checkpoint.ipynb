# Histplot

A histogram is a classic visualization tool that represents the distribution of one or more variables by counting the number of observations that fall within disrete bins.

This function can normalize the statistic computed within each bin to estimate frequency, density or probability mass, and it can add a smooth curve obtained using a kernel density estimate

**seaborn.histplot**(data=None,x=None, y=None, hue=None, bins='auto', kde=False)

```python
sns.__version__
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguins.shape
```

```python
penguins.isnull().sum()
```

```python
penguins.dropna(inplace=True)
```

```python
penguins.shape
```

## Basics

```python
sns.set_style("darkgrid")
```

```python
# drawing basic histplot using seaborn

sns.histplot(penguins.flipper_length_mm)
```

```python
sns.histplot(penguins.bill_length_mm)  # you can pass the column name together with df name
```

```python
# you can pass both df name as keyword argument and column name as x parameter

sns.histplot(data= penguins, x= "bill_length_mm")
```

```python
# to produce horizontal histogram use y parameter instead of x parameter

sns.histplot(data= penguins, y= "bill_length_mm")  
```

```python
#Adding kde(kernel density estimation) curve

sns.histplot(data= penguins, x= "bill_length_mm", kde=True)
```

```python
# Adjusting special bins location

sns.histplot(data= penguins, x= "bill_length_mm", bins=[25,30, 35, 40, 50,53,55,60])
```

## binwidth and binrange

```python
# you can adjust binwidth(width of each bar) and binrange(start value, stop value) using binwidth and binrange parameters
sns.histplot(data= penguins, x= "bill_length_mm", binwidth=5, binrange=(30,60))
```

## Stat

```python
# You can plot some statistical calculation results using stat parameter with sns.histplot()

sns.histplot(data= penguins, x= "bill_length_mm", stat="count")  # count of bill_lengths
```

```python
sns.histplot(data= penguins, x= "bill_length_mm", stat="frequency")  # frequency of bill_lengths
```

```python
# density : AREA of histogram sums is equal to one(1)

sns.histplot(data= penguins, x= "bill_length_mm", stat="density")  # density of bill_lengths
```

```python
# probability : HEİGHT of histogram bars is equal to one(1)

sns.histplot(data= penguins, x= "bill_length_mm", stat="probability")
```

```python
# cumulative plot is represents sum of probability heihgts. It's end point is equal to one

sns.histplot(data= penguins, x= "bill_length_mm", stat="probability", fill=False, element="step")
```

```python
# cumulative plot is represents sum of probability heihgts. It's end point is equal to one

sns.histplot(data= penguins, x= "bill_length_mm", stat="probability", fill=False, element="step", cumulative=True)
```

## Categorical Variables

```python
# drawing bill_lengt histogram according to different species

sns.histplot(data= penguins, x= "bill_length_mm", hue="species");

# Different colors for each different species was set
# Some bars of different species is overlapping
```

```python
# to prevent overlapping there are two ways
# 1. using element parameter as step

sns.histplot(data= penguins, x= "bill_length_mm", hue="species", element="step");
```

```python
# to prevent overlapping there are two ways
# 2. using element parameter as ploy (to draw ploygans for each categories)

sns.histplot(data= penguins, x= "bill_length_mm", hue="species", element="poly");

# now it is easy to distinguish each category layer
```

## multiple

```python
# Drawing stacked histogram using multiple parameters as stack

sns.histplot(data= penguins, x= "bill_length_mm", hue="species", multiple="stack");
```

```python
# to draw proportion of each category with histogram, we use multiple parameter as fill

sns.histplot(data= penguins, x= "bill_length_mm", hue="species", multiple="fill");
```

## Bivariate Histplot

```python
# to draw bivariate histplot we use x and y parameters together

sns.histplot(x="bill_length_mm", y= "bill_depth_mm", data=penguins);
```

### hue parameter

```python
# we can add categorical variables to the bivariate histplot using hue parameter

sns.histplot(x="bill_length_mm", y= "bill_depth_mm", data=penguins, hue="species");

# overlappings are not seen exactly
```

```python
# we can use the same categorical variable as y and hue parameter of the bivariate histplot
# then we can get more meaningful histplot

sns.histplot(x="bill_length_mm", y= "species", data=penguins, hue="species");
```

## Styling

### shrink

```python
# to adjust the distansce between the species bars we use shrink parameter
# to draw bars side by side according to hue parameter we use multiple="dodge" parameter

sns.histplot(x="species", hue= "sex", multiple="dodge", data=penguins, shrink=0.8);
```

## color, palette parameter

```python
# adjusting bar colors using color parameter

sns.histplot(x="bill_length_mm", data=penguins, color="indigo");
```

```python
# adjusting bar colors of bivariate histogram using palette parameters

sns.histplot(x="species", hue= "sex", multiple="dodge", data=penguins, shrink=0.8, palette="bone");
```

## fill parameter

```python
# to draw histplot bars only with outlines we use fill parameter as False
sns.histplot(x="bill_length_mm", data=penguins, color="indigo", fill=False);
```

# Displot

**seaborn.displot**(data=None, x=None, y=None, hue=None, rug=False)

This function provides access to several approaches for visualizing the univariate or bivariate distribution of data, including subsets of data defined by semantic mapping and faceting across multiple subplots

```python
cars = sns.load_dataset("mpg").dropna()
cars.head()
```

## Basics

```python
sns.set_style("white")
```

```python
sns.displot(x="weight", data=cars);  # count distribution plot
```

```python
sns.displot(x="weight", data=cars, kde=True);  # distribution + kde(kernel density estimation) plot
```

```python
sns.displot(x="weight", data=cars, kind="ecdf");  # ecdf: emprical cumulative distribution function
```

```python
sns.displot(x="weight", data=cars, kind="kde", rug=True); # rug plot
```

## Categorical variables: hue parameter

```python
sns.displot(x="weight", data=cars, hue= "origin");  # displot of cars according to origin. There are so many overlapping
```

```python
sns.displot(x="weight", data=cars, hue= "origin", kind="kde"); # it is better to display each origin using kind: "kde" plot
```

## Bivariate (2D) Options

```python
sns.displot(x="weight", y="horsepower", data=cars); # two dimensional distribution plot

# dark areas represent bigger density
```

```python
sns.displot(x="weight", y="horsepower", data=cars, kind="kde", rug=True);  # kde plot with rug parameter(shows density in each axis)
```

## Small Multiples (FacetGrid)

```python
sns.displot(x="weight", data=cars, hue= "origin", col="origin"); # each origin was plotted in different columns 
```

```python
sns.displot(x="weight", data=cars, hue= "origin", col="origin", kind="kde", rug=True);

# kde plot with rug was plotted in different columns using col parameter
```

```python
sns.displot(x="weight", data=cars, col="origin");  # distribution of weight for each origin in different columns
```

```python
sns.displot(x="weight", data=cars[cars.cylinders.isin([4,6,8])],
            col="origin",
            row="cylinders",
            hue="cylinders");
```

## Styling

### palette

```python
sns.displot(x="weight", data=cars, hue= "origin", col="origin", palette="crest");  # to change colors of bars palette parameter is used 
```

### height, aspect

```python
sns.displot(x="weight", data=cars, col="origin", height=2); # height default value is 5. For smaller charts set height less than 5 
```

```python
sns.displot(x="weight", data=cars, col="origin", height=2, aspect=1.5);  # aspect measuser width divided by height
```

```python
# bins parameter is used for displot also

sns.displot(x="weight", data=cars, bins=20);
```

```python
# we can also use fill parameter for kde plot. line width also can be changed with lw parameter
sns.displot(x="weight", data=cars, kind="kde", fill=True, lw=3);
```

```python
# we can set some styles for rug parameter as rug_kws
sns.displot(x="weight", data=cars, kind="kde", rug=True, rug_kws={"height" : 0.05, "color" : "r"});
```

## FacetGrid Object

### facet_kws

```python
sns.displot(x="weight", data=cars, col="origin", facet_kws={"despine": False});
```

```python
g = sns.displot(x="weight", data=cars, col="origin", facet_kws={"despine": False});  # we created a FacetGrid object as g
```

```python
type(g)  # its' type is
```

```python
g.axes_dict.items()  # these items acan be used to adjust individualy
```

```python
g = sns.displot(x="weight", data=cars, col="origin")

for col_val, ax in g.axes_dict.items():
    if col_val == "europe":
        ax.set_facecolor("#E2F2FF");
    
```
