### Functional Method

````python
y1 = x
y2 = x ** 2
y3 = x ** 3
y4 = np.sqrt(x)

plt.figure()  # Create a new figure, or activate an existing figure.
plt.subplot(2,2,1)  # plt.subplot(nrows, ncols, index, **kwargs)
plt.plot(x,y1,"ro")
plt.title("$y_1=x$")


plt.subplot(2,2,2)
plt.plot(x,y2,"g--")
plt.title("$y_2=x^2$")

plt.subplot(2,2,3)
plt.plot(x,y3,"b^")
plt.title("$y_3=x^3$")

plt.subplot(2,2,4)
plt.plot(x,y4,"ks")
plt.title("$y_4=\sqrt{x}$")

plt.tight_layout()

# we placed 4 graphics inside a figure.
````

````python

plt.figure()  # Create a new figure, or activate an existing figure.
plt.subplot(4,1,1)  # plt.subplot(nrows, ncols, index, **kwargs)
plt.plot(x,y1,"ro")
plt.title("$y_1=x$")


plt.subplot(4,1,2)
plt.plot(x,y2,"g--")
plt.title("$y_2=x^2$")

plt.subplot(4,1,3)
plt.plot(x,y3,"b^")
plt.title("$y_3=x^3$")

plt.subplot(4,1,4)
plt.plot(x,y4,"ks")
plt.title("$y_4=\sqrt{x}$")  # Latex Typing

plt.tight_layout()
````

###  Bar Chart:
```matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, align='center', data=None)```

Make a bar plot.
The bars are positioned at x with the given alignment. Their dimensions are given by height and width. 

Many parameters can take either a single value applying to all bars or a sequence of values, one for each bar.

The vertical baseline is bottom (default 0).

### Histogram
```matplotlib.pyplot.hist(x, bins=None, range=None, density=False, orientation='vertical', color=None, label=None, stacked=False,data=None)```

Plot a histogram.
Compute and draw the histogram of x. The return value is a tuple (n, bins, patches) or ([n0, n1, ...], 
Multiple data can be provided via x as a list of datasets of potentially different length ([x0, x1, ...]), 
Masked arrays are not supported.
The bins, range, weights, and density parameters behave as in numpy.histogram.
or as a 2D ndarray in which each column is a dataset. 
Note that the ndarray form is transposed relative to the list form.
bins, [patches0, patches1, ...]) if the input contains multiple data. 
See the documentation of the weights parameter to draw a histogram of already-binned data.


> Histograms are bar charts where the y-axis contains the frequency of a variable.

````python
# all the sugar numbers for different patients in this list.
# when you are working with a histogram you need only single dim. array the axis-y contain the frequency of those different numbers to plot hist.
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar,bins=3,rwidth=0.95)
````

````python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color="g")  # histtype="step"
````

````python
blood_sugar_men= [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,150,97,119,79,89,67,112,139]
plt.xlabel("sugar range")
plt.ylabel("Total no of patients")
plt.title("Blood sugar analysis")
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=["green","orange"],
        label=["men","women"])
plt.legend()
````

### Pie chart
```  matplotlib.pyplot.pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, labeldistance=1.1, startangle=0, radius=1, data=None)```

Plot a pie chart.
Make a pie chart of array x. The fractional area of each wedge is given by x/sum(x). 
The wedges are plotted counterclockwise, by default starting from the x-axis.
If sum(x) < 1, then the values of x give the fractional area directly and the array will not be normalized. 
The resulting pie will have an empty wedge of size 1 - sum(x).
