## Matplotlib Figure Object

### Matplotlib Object Oriented Method

This means we will create a figure object and then call methods or attributes from that object.


```python
# Matplotlib object oriented metod ile öce bir figure objecti oluştururuz.
# Bu objectin ardından yöntemleri ve özellikleri çağırarak grafiklerimizi oluştururuz.
```

```python
a = np.linspace(0,10,11)
b = a ** 4
a, b
```

```python
x = np.arange(0,10)
y = 2*x
x, y
```

### Creating a Figure

The main idea in using the more formal Object Oriented method is to create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it.

```python
# Creates blank canvas

fig = plt.figure()
fig
```

```python
plt.figure(figsize=(10,10))  # figure objectinin büyüklüğünü ayarlamak için 10 inch
fig
```

### NOTE: All the commands need to go in the same cell!

To begin we create a figure instance. Then we can add axes to that figure:

```python
# Create Figure (empty canvas)

fig =plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0,0,1,1])  # left(x coordinate), bottom(y coordinate), width, height (range 0 to 1)

# plot on that set of axes
axes.plot(x,y)

plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b)
plt.show()
```

### Adding another set of axes to the figure

So far we've only seen one set of axes on this figure object, but we can keep adding new axes on to it at any location and size we want. We can then plot on that new set of axes.

```python
type(fig) # fig türüne baktığımızda figure objecti olduğunu görebiliriz.
```

```python
fig = plt.figure()  # to create empty canvas

axes1 = fig.add_axes([0,0,1,1])  # axes of large figure
axes2 = fig.add_axes([0.2, 0.4, 0.4, 0.4])  # axes of small figure

axes1.plot(a,b)  # plot of large figure 

axes1.set_xlabel("X label")  # x label of large figure
axes1.set_ylabel("Y label")  # y label of large figure
axes1.set_title("Large Figure")  # title of large figure


axes2.plot(a,b)  # plot of small figure

axes2.set_xlabel("X label")  # x label of small figure
axes2.set_ylabel("Y label")  # y label of small figure
axes2.set_title("Small Figure")  # title label of small figure

plt.show()
```

```python
fig = plt.figure()  # to create empty canvas

axes1 = fig.add_axes([0,0,1,1])  # axes of large figure
axes2 = fig.add_axes([0.2, 0.4, 0.4, 0.4])  # axes of small figure

axes1.plot(a,b)  # plot of large figure 

axes1.set_xlabel("X label")  # x label of large figure
axes1.set_ylabel("Y label")  # y label of large figure
axes1.set_title("Large Figure")  # title of large figure


axes2.plot(a,b)  # plot of small figure

axes2.set_xlim(1,5)  # x limits(start, end) of small figure
axes2.set_ylim(0,700)  # y limits(start, end) of small figure
axes2.set_xlabel("X label")  # x label of small figure
axes2.set_ylabel("Y label")  # y label of small figure
axes2.set_title("Small Figure")  # title label of small figure

plt.show()
```

## Linestyles

```python
# possible linestype options ‘--‘, ‘–’, ‘-.’, ‘:’, ‘steps’

fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x-1, color="green", lw=3, linestyle="-")
ax.plot(x, x-2, color="blue", lw=3, ls="-.")
ax.plot(x, x-3, color="red", lw=3, ls=":")
ax.plot(x, x-4, color="yellow", lw=3, ls="--");
```

### Custom linestyle dash
The dash sequence is a sequence of floats of even length describing the length of dashes and spaces in points.

For example, (5, 2, 1, 2) describes a sequence of 5 point and 1 point dashes separated by 2 point spaces.

First, we see we can actually "grab" the line from the .plot() command

```python
# adjusting dash style

fig, ax = plt.subplots(figsize=(12,6))

lines = ax.plot(x, x+1, color="purple", lw=5, linestyle="-")

lines[0].set_dashes([1, 2, 7, 2]); # lines değişkeni bir liste olduğu için ilk elemanını seçerek onun dash sitilini düzenledik.

# [1, 2, 7, 2]: 1 nokta, 2 boşluk, 7 nokta, 2 boşluk
```

```python
# adjusting dash style

fig, ax = plt.subplots(figsize=(12,6))

lines = ax.plot(x, x+1, color="purple", lw=5, linestyle="-")

lines[0].set_dashes([1, 2, 1, 2, 10, 2]); # lines değişkeni bir liste olduğu için ilk elemanını seçerek onun dash sitilini düzenledik.

# [1, 2, 1,2, 10, 2]: 1 nokta, 2 boşluk, 1 nokta, 2 boşluk, 10 nokta, 2 boşluk
```
