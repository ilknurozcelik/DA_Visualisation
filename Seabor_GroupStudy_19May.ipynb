```python
plt.figure(figsize=(12,8))
sns.histplot(data=df[(df["salary"]>90000)&(df["salary"]<100000)], x="salary", 
             hue="division",multiple="dodge",shrink=0.5,bins=5)
```

```python
plt.figure(figsize=(12,8))
ax =sns.histplot(data=df[(df["salary"]>90000)&(df["salary"]<100000)], x="work experience", 
             hue="division",multiple="dodge",shrink=0.5,bins=5)
sns.move_legend(ax, loc="center", bbox_to_anchor=(1.2,0.5), title=None, frameon=False)
```

```python
plt.figure(figsize=(12,8))
ax = sns.histplot(data=df[(df["salary"]>90000)&(df["salary"]<100000)], x="work experience", 
             hue="division",multiple="dodge",shrink=0.5,bins=5)

sns.move_legend(obj=ax, loc="center", bbox_to_anchor=(1.2, 0.6),title="LEGEND TİTLE", frameon=True)

for i in range(len(ax.containers)):
    ax.bar_label(ax.containers[i],label_type="edge",padding=0.2)
plt.show()
```
